```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Uts_Big_Data_Processing") \
    .getOrCreate()
sc = spark.sparkContext
```

```python
!wget https://raw.githubusercontent.com/dwizz11/UTS-Big-Data-Processing/main/Table1.csv
```

```python
data1 = "Table1.csv"
```

```python
dataframe1=spark.read.option("delimiter", ",").option("header", "true").csv(data1)
```

```python
dataframe1.show()
```

```python
!wget https://raw.githubusercontent.com/dwizz11/UTS-Big-Data-Processing/main/Table2.csv
```

```python
data2 = "Table2.csv"
```

```python
dataframe2=spark.read.option("delimiter", ",").option("header", "true").csv(data2)
```

```python
dataframe2.show()
```

```python
dataframe1.createOrReplaceTempView("DATAFRAME1")
dataframe2.createOrReplaceTempView("DATAFRAME2")
```

```python
Result=spark.sql("select t1.user_id_sender as user_id_sender,t2.age as age_sender,t2.gender as gender_sender,t1.user_id_recipient as user_id_recipient,t3.age as age_recipient,t3.gender as gender_recipient,t1.image_id as image_id,t1.like from DATAFRAME1 t1 , DATAFRAME2 t2, DATAFRAME2 t3 where t1.user_id_sender==t2.user and t3.user==t1.user_id_recipient")
```

```python
Result.show()
```
